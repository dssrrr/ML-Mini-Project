# 1. List all the cities and the respective number of pending orders in 'India'?

# 2.Which country has the most suspected fraud?

```python
l1=df[df['Order Status']=='SUSPECTED_FRAUD']
```

```python
a=l1.groupby(['Order Country','Order Status']).size().reset_index(name='fraud_count')
```

```python
c=a[a['fraud_count']==a['fraud_count'].max()]
```

# 4. Which year had the minimum sales for "Nike" products?

```python
l=[]
for i in df['Product Name']:
    if(i[0:4]=='Nike'):
        l.append(i)
```

```python
df4=df[df['Product Name'].isin(l)]
```

```python
df4['Product Name'].value_counts()
```

```python
sales=df4.groupby(['shipping years'])['Sales'].sum().reset_index(name='tot_sales')
```

```python
sales
```

In the year=2018 there are no Sales for Nike Product

```python
sales[sales['tot_sales']==sales['tot_sales'].min()]
```

```python
sales['tot_sales'].min()
```
